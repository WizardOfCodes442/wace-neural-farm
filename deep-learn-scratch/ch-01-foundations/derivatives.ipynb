```python
#derivatives like function are a very important concept
#for understanding deep learning
#also like a function, derivatives can be depicted in multiple ways
#at a high level, the derivative of a function at a point is the rate of change
#of the output of the function with respect to it's input at that point
#working through the three concept of representation

"""
first we'll get mathematical precise:
we can describe this number how much the output of f
changes as we change its input at a particular valuye a of the input -
as a limit
    df/ (a) = lim  (f(a+ h) - f(a- h)) /
      du      h->0                     2*a

  This limit can be approximated numerically by setting a very
  small value for h. such as 0.0001, so we can compute
  derivative as:

    df/ (a) = (f(a + 0.0001) - f(a - 0.0001) ) /
      du                                     0.002

"""
#Diagrams

"""
First the familiar way
If we draw a tangent line to the cartesian

|         (
|           (
|------------\(
|-------------\(
|--------------\(
|             |
|             |
|_____________|____________________________
"""

#thinking of functions as minifactories
#think of inputs to those facories being
#connected to the output of a string

#code
from typing import Callable
from numpy import ndarray
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def f(x):
    return x**2  # Example function: x^2

# Define the derivative of the function
def df(x):
    return 2*x  # Derivative of x^2

# Point where we want to find the tangent
x0 = 2

# Calculate the tangent line
y0 = f(x0)
m = df(x0)  # Slope of the tangent
tangent_line = lambda x: m*(x - x0) + y0

# Generate x values for the plot
x = np.linspace(0, 4, 100)  # Adjust the range as needed

# Plot the curve and the tangent line
plt.plot(x, f(x), label='Curve')
plt.plot(x, tangent_line(x), label='Tangent')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curve and Tangent')

# Add grid for better visualization
plt.grid(True)
# Add Legend
plt.legend()
# Display the plot
plt.show()

# Define the function (flipped to face right)
def f(x):
    return np.sqrt(x)  # Example: Square root function

# Define the derivative of the function
def df(x):
    return 1 / (2 * np.sqrt(x))  # Derivative of sqrt(x)

# Point where we want to find the tangent
x0 = 2

# Calculate the tangent line
y0 = f(x0)
m = df(x0)
tangent_line = lambda x: m * (x - x0) + y0

# Generate x values for the plot
x = np.linspace(0, 4, 100)

# Plot the curve and the tangent line (flipped)
plt.plot(f(x), x, label='Curve')  # Flipped: x and f(x) swapped
plt.plot(tangent_line(x), x, label='Tangent')  # Flipped: x and tangent_line(x) swapped

# Add dotted lines to x and y axes
plt.axhline(0, color='black', linestyle='dotted')  # Horizontal line at y=0
plt.axvline(0, color='black', linestyle='dotted')  # Vertical line at x=0

# Add labels and title
plt.xlabel('y')  # Swapped labels
plt.ylabel('x')  # Swapped labels
plt.title('Curve and Tangent (Flipped)')

# Add grid and legend
plt.grid(True)
plt.legend()

# Display the plot
plt.show()

def deriv(func: Callable[[ndarray], ndarray],
          input_: ndarray, delta: float = 0.001) -> ndarray:
    return (func(input_ + delta) - func(input_ - delta)) / (2 * delta)
```
